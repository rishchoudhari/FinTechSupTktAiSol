```python
billing_examples = [
    "I was charged twice this month.",
    "Why is my bill higher than usual?",
    "There is an unknown charge on my account.",
]

failed_payment_examples = [
    "My payment keeps getting declined.",
    "The transfer failed again.",
    "I can't complete my payment on your platform.",
]

refund_examples = [
    "I still haven't received my refund.",
    "How long does a refund take?",
    "Please process my refund request.",
]

account_examples = [
    "I need to update my bank details.",
    "I can't verify my identity.",
    "My account is locked and I can't log in.",
]

technical_examples = [
    "The app crashes when I open it.",
    "The website won't load.",
    "I keep getting an error message.",
]

general_examples = [
    "How do I change my email?",
    "Where can I find my settings?",
    "What are your support hours?",
]

def expand(messages, label, priority):
    return [{"text": m, "label": label, "priority": priority} for m in messages]

dataset = (
    expand(billing_examples, "Billing issue", "High") +
    expand(failed_payment_examples, "Failed payment", "High") +
    expand(refund_examples, "Refund request", "Medium") +
    expand(account_examples, "Account access / KYC", "Medium") +
    expand(technical_examples, "Technical issue", "High") +
    expand(general_examples, "General question", "Low")
)

df = pd.DataFrame(dataset)
df.sample(5)
```

```python
df.to_csv("support_tickets.csv", index=False)
```
